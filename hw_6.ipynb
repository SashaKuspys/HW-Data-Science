```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

```python
dataset = pd.read_csv('data_2d.csv')
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(dataset)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Кількість кластерів')
plt.ylabel('WCSS')
plt.show()
```

```python
df = pd.read_csv('data_2d.csv')
k_means = KMeans(n_clusters=2)
k_means.fit(df)

plt.scatter(df.iloc[:, 1] , df.iloc[:, 2], c=k_means.labels_)
plt.show()
```

```python
dataset_1 = pd.read_csv('mnist.csv')
dataset_1
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import minmax_scale

data = minmax_scale(dataset_1)

X_reduced = PCA(n_components=2).fit_transform(data)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1])
plt.show()
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_reduced)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel('Кількість кластерів')
plt.ylabel('WCSS')
plt.show()
```

```python
k_means = KMeans(n_clusters=3)
k_means.fit(X_reduced)

plt.scatter(X_reduced[:, 0] , X_reduced[:, 1], c=k_means.labels_)
plt.show()
```
