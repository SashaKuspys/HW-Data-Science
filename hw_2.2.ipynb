10) Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

12) Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_data = data.groupby('Посада')

print(grouped_data.size())
```

13) Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
aggregated_data = data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})

aggregated_data
```

14) Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    avg_salary = row.mean()
    return avg_salary

aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)

aggregated_data
```

15) Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
avg_stats = aggregated_data['avg'].describe()
avg_stats
```

16) Збережіть отриману таблицю в CSV файл

```python
avg_stats.to_csv('avg_stats.csv', header=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['Зарплата на місяць'], bins=20, color='skyblue', edgecolor='black')
plt.title('Розподіл зарплат за місяць')
plt.xlabel('Зарплата, грн')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
language_counts = data['Мова програмування'].value_counts()
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Розподіл мов програмування')
plt.axis('equal')  # Забезпечує кругову форму діаграми
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([data[data['Мова програмування'] == 'Python']['Зарплата на місяць'],
             data[data['Мова програмування'] == 'Java']['Зарплата на місяць'],
             data[data['Мова програмування'] == 'C++']['Зарплата на місяць']],
            labels=['Python', 'Java', 'C++'])
plt.title('Розподіл зарплат за мовами програмування')
plt.xlabel('Мова програмування')
plt.ylabel('Зарплата, грн')
plt.grid(True)
plt.show()
```
