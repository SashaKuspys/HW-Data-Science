Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання

```python
match='Коефіцієнт народжуваності в регіонах України'
table = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',thousands="",
                          decimal=",",
                          match = match,
                    )
table
```

3) Замініть у таблиці значення "—" на значення NaN

```python
tabl = table[0]

df = pd.DataFrame(tabl)

df.replace("—", float("NaN"), inplace=True)

df
```

4) Визначте типи всіх стовпців за допомогою dataframe.dtypes

5) Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
for column in df.columns:
    if column != 'Регіон' and df[column].dtype == 'object':
        df[column] = pd.to_numeric(df[column], errors='coerce')

print(df.dtypes)
```

6) Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

7) Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = table[0]

tab1 = df.drop(df.index[-1], inplace=True)

print(tabl)
```

8) Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
import numpy as np

tab1 = df
tab1.replace("—", np.nan, inplace=True)

for col in tab1.columns[1:]:
    if tab1[col].dtype == "object":
        tab1[col] = pd.to_numeric(tab1[col], errors="coerce")

print(tab1.dtypes)

tab1 = tab1.fillna(tab1.mean(numeric_only=True))
tab1
```

9) Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python

average_birth_rate_2019_ukraine = tabl['2019'].mean()
regions_with_higher_birth_rate = tabl[tabl['2019'] > average_birth_rate_2019_ukraine]['Регіон']
regions_with_higher_birth_rate
```

10) У якому регіоні була найвища народжуваність у 2014 році?

```python
index_max_birth_rate_2014 = tabl['2014'].idxmax()

region_with_max_birth_rate_2014 = tabl.loc[index_max_birth_rate_2014, 'Регіон']

print("Регіон з найвищою народжуваністю у 2014 році:", region_with_max_birth_rate_2014)
```

11) Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt

birth_rate_2019 = tab1['2019']

regions = tab1['Регіон']

plt.figure(figsize=(10, 6))
plt.barh(regions, birth_rate_2019, color='skyblue')
plt.xlabel('Народжуваність у 2019 році')
plt.ylabel('Регіон')
plt.title('Народжуваність по регіонах у 2019 році')
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.hist(birth_rate_2019, bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Народжуваність')
plt.ylabel('Частота')
plt.title('Гістограма народжуваності по регіонах у 2019 році')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(birth_rate_2019, labels=regions, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Розподіл народжуваності по регіонах у 2019 році')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(regions, birth_rate_2019, marker='o', color='orange', linestyle='-')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Залежність народжуваності від регіону у 2019 році')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
